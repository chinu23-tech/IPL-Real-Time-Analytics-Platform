```python
import pandas as pd
import time
```

```python
df = pd.read_csv("../data/IPL_fixed.csv")
print(df.shape)
```

```python
match_id = df['match_id'].iloc[0]
match_df = df[df['match_id'] == match_id]
print("Match ID:", match_id)
print("Balls:", len(match_df))
```

```python
for _, row in match_df.head(20).iterrows():
    print(
        f"Over {row['over']}.{row['ball']} | "
        f"{row['batting_team']} | "
        f"Runs: {row['runs_total']}"
    )

    time.sleep(1)
```

```python
for _, row in match_df.iterrows():
    print(
        f"Over {row['over']}.{row['ball']} | "
        f"{row['batting_team']} | "
        f"Runs: {row['runs_total']}"
    )

    time.sleep(0.1)
```
