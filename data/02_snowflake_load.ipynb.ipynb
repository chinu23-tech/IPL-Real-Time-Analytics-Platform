```python
import pandas as pd

df = pd.read_csv("../data/IPL_snowflake.csv")

bad_rows = df[
    pd.to_numeric(df['event_match_no'], errors='coerce').isna()
]

print("Bad rows:", len(bad_rows))

bad_rows[['event_match_no']].drop_duplicates().head(20)
```
